### Cargar datos

```python
contenido = pd.read_csv('starwars.csv')
print(contenido.keys())
    
columnas_a_usar = ['name','gender','homeworld','species']

df = pd.read_csv(
        'starwars.csv',
        usecols=columnas_a_usar)
```

# Genero de personajes

```python
generos = df.groupby('gender')
tipo_generos = []
total_generos_personajes = []


for genero,personaje in generos:
    print(genero)    
    print(len(personaje))
    tipo_generos.append(genero)
    total_generos_personajes.append(len(personaje)) 
 

plt.pie(total_generos_personajes,labels=tipo_generos,
autopct='%1.2f%%')
plt.legend(tipo_generos)
plt.axis('equal')
plt.show()
    
```

### Estatura maxima de cada especie

```python
estaturas_Por_Especie = contenido.groupby('species')['height']
todas_estaturas= []
todas_especies = []
maximo_por_especie=[]



for especie, height in estaturas_Por_Especie:
   #print(especies)
   todas_especies.append(especie)
   maximo_por_especie.append(max(height))

#print(maximo_por_especie)
   
#print(todas_especies)


index = np.arange(len(todas_especies)) 
plt.bar(index,maximo_por_especie)   
plt.xlabel('todas_especies')
plt.ylabel('maximo_por_especie')

plt.xticks(index,todas_especies,rotation=90)
 
plt.show()
   
  
    
    

    
    
```

### Estatura minima por especie 

```python
estaturas_Por_Especie = contenido.groupby('species')['height']
todas_estaturas= []
todas_especies = []
minimo_por_especie=[]



for especie, height in estaturas_Por_Especie:
   #print(especies)
   todas_especies.append(especie)
   minimo_por_especie.append(min(height))

#print(minimo_por_especie)
   
#print(todas_especies)


index = np.arange(len(todas_especies)) 
plt.bar(index,minimo_por_especie)   
plt.xlabel('todas_especies')
plt.ylabel('minimo_por_especie')

plt.xticks(index,todas_especies,rotation=90)
 
plt.show()
   
  
```
